# Introdução

### Por que estudar aposentados?
   Estudar os aposentados é um fator importante de ser estudado, já que atualmente o número de jovens atualmente é muito alto e estes serão os aposentados do futuro, além disso a natalidade infantil abaixou muito, o que fará a quantidade de aposentados serem muito maior que a taxa de jovens, o que será um grande problema. Por isso uma pesquisa sobre os aposentados é interessante para podermos ter uma estimativa do que pode fazer a expectativa de vida aumentar.

Os dados que eu pretendo obter não vai ser possível comparar, talvez existam dados, mas o meu trabalho tem objetivo em criar, por exemplo, um plano de seguro para idosos e ver como se deve fazer planos, então são informações que uma pessoa que vá criar um produto para aposentados observarem o comportamento.

### Escolha dos dados:

A escolha dos dados primeiramente foi pensando em ver comportamentos que façam a expectativa de vida aumentar, porém como não há uma taxa de mortalidade, não pude comprovar nada, então comecei a analisar dados que mostram a maior parcela de idosos, relacionando sexo, raça, local da moradia,idade de primeiro trabalho e se sabe ler e escrever, além da idade para criar bases de comparação.

## Gráficos um pouco mais complexos

Usando a tabela cruzada temos uma base melhor de comparação entre grupos e possuirmos uma base melhor para concluir o que esperamos, é essencial fazer tabelas cruzadas possuindo um propósito de uma resposte, de uma conclusão que se quer obter atraves das informações.

```python
fig, axs = plt.subplots(2,2,figsize=(12, 15))
sexo_idade92 = pd.crosstab(idade1992,Grupo1992.v0302)
sexo_idade01 = pd.crosstab(idade2001,Grupo2001.v0302)
sexo_idade11 = pd.crosstab(idade2011,Grupo2011.V0302)
sexo_idade14 = pd.crosstab(idade2014,Grupo2014.V0302)

sexo_idade92.plot(ax=axs[0,0])
axs[0,0].set_xlabel('Faixa de Idade',fontsize=14)
axs[0,0].set_ylabel('Frequencia absoluta do sexo',fontsize=14)
axs[0,0].set_title('1992', fontsize=26)
sexo_idade01.plot(ax=axs[0,1], legend=False)
axs[0,1].set_xlabel('Faixa de Idade',fontsize=14)
axs[0,1].set_ylabel('Frequencia absoluta do sexo',fontsize=14)
axs[0,1].set_title('2001', fontsize=26)
sexo_idade11.plot(ax=axs[1,0], legend=False)
axs[1,0].set_xlabel('Faixa de Idade',fontsize=14)
axs[1,0].set_ylabel('Frequencia absoluta do sexo',fontsize=14)
axs[1,0].set_title('2011', fontsize=26)
sexo_idade14.plot(ax=axs[1,1], legend=False)
axs[1,1].set_xlabel('Faixa de Idade',fontsize=14)
axs[1,1].set_ylabel('Frequencia absoluta do sexo',fontsize=14)
axs[1,1].set_title('2014', fontsize=26)
```
