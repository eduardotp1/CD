### Distribuição Teórica

<strong>Normal:</strong> Esta distribuição teórica mostra uma simetria entre as proporções de pessoas ao longo da hora, o que é possível observar no gráfico acima.

<strong>Poisson:</strong> A distribuição teórica de poisson encaixa no tema proposto, já que uma venda não é gerado devido uma venda anterior, a questão da venda seria um caso aleatório. O que na prática, é o que ocorre.

### 5)

### Gráficos com  modelos de distribuição teórica.

Utilizando o comando stats.norm, podemos fazer a estimativa de como ficaria a distribuição do gráfico em um comportamento normal.

Apesar de ter mencionado anteriormente que usaria a distribuição de Poisson, decidi alterar para a distribuição semicircular, já que é uma distribuição nao usada ainda. Utilizando o stats.semicircular, pois em um exemplo do site do scipy.stats, o modelo aparenta igual ao estudado.

```python
mu = (dados.quant8_9.median())
sigma = (dados.quant8_9.var())**0.5
x=sorted(stats.norm.rvs(size=1000, loc=mu, scale=sigma))
plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), lw=3, alpha=0.6,label="Normal")
plt.plot(x, stats.semicircular.pdf(x,loc=mu, scale=sigma), lw=3, alpha=0.6, color="red",label="Semicircular")
dados.quant8_9.plot.hist(bins=faixa, normed=1)
plt.title('Gráfico com Distribuição Teórica', fontsize=20)
plt.legend()
```

Podemos observar que o comportamento da distribuição normal é muito pareciso com o gráfico. E em termos de formato, o semicircular não se difere muito também. Mas, entre elas, a melhor escolha seria a da normal.

### 6)

```python
nfit=stats.norm.fit(x)
cfit=stats.semicircular.fit(x)
plt.plot(x, stats.norm.pdf(x, loc=nfit[0], scale=nfit[1]), lw=3, alpha=0.6,label="Normal")
plt.plot(x, stats.semicircular.pdf(x,loc=cfit[0], scale=cfit[1]), lw=3, alpha=0.6, color="red",label="Semicircular")
dados.quant8_9.plot.hist(bins=faixa, normed=1)
plt.title('Gráfico com Distribuição Teórica e Fit', fontsize=20)
plt.legend()
```

## Q-Q Plot

```python
quant8_9=sorted(dados.quant8_9)
freq_acum = [] 
for i in range(len(dados.quant8_9)):
    acum = (i-0.5 ) / len(dados.quant8_9)
    if acum < 0:
        acum = 0
    freq_acum.append(acum)

freq_acum = pd.Series(data=freq_acum)

# Gera os valores dos quantis teóricos para as distribuições analisadas
q_normal = stats.norm.ppf(freq_acum, nfit[0], nfit[1]) # quantis teóricos de alpha
q_semicircular = stats.semicircular.ppf(freq_acum,  cfit[0], cfit[1]) # quantis teóricos de gamma


plt.title("Q-Q plot: Dados x Normal x Semicircular", fontsize=20)

normal_plot, = plt.plot(quant8_9, q_normal, 'o', label="Normal")
semi_plot, = plt.plot(quant8_9, q_semicircular, 'o', color="red", label = "semicircular")

plt.legend(handles=[normal_plot, semi_plot])
plt.plot((min(quant8_9), max(quant8_9)), (min(quant8_9), max(quant8_9)), '-')

for i in range(len(quant8_9)):
    plt.plot((quant8_9[i], quant8_9[i]), (q_normal[i], quant8_9[i]), '-', color='red', alpha=0.3  )
    plt.plot((quant8_9[i], quant8_9[i]), (q_semicircular[i], quant8_9[i]), '-', color='green', alpha=0.3  )
```

### 9)

```python
cdf_normal = stats.norm.cdf(quant8_9, nfit[0], nfit[1])
cdf_semicircular = stats.semicircular.cdf(quant8_9,  cfit[0], cfit[1])


quant8_9_plot, = plt.plot(quant8_9, freq_acum, "-", label="Dados")
normal_plot, = plt.plot(q_normal, cdf_normal, 'o', color="black", label="Normal")
semi_plot, = plt.plot(q_semicircular, cdf_semicircular, 'o', color="blueviolet", label = "Semicircular")
plt.legend(handles=[quant8_9_plot,normal_plot, semi_plot],loc='best')
plt.xlabel("FDA")
plt.ylabel("FRA")
plt.plot()
```

O gráfico da normal acompanha melhor os dados, portanto, a escolha feita anteeriormente pode ser concluída, com o gráfico acima; para os dados escolhidos, a melhor escolha seria a <strong>normal</strong> e a semicircular não foi a melhor opção.

### Erro quadrático quantil

```python
erro_quadratico_normal=0
erro_quadratico_semi=0
for i in range(1,len(q_normal)-1):
    erro_quadratico_normal += ((quant8_9[i] - q_normal[i])**2)
for k in range(1,len(q_semicircular)-1):
    erro_quadratico_semi += ((quant8_9[k] - q_semicircular[k])**2)
print("Erro quadrático da normal vs FRA: ", (erro_quadratico_normal))
print("Erro quadrático da semicircular vs FRA: ", (erro_quadratico_semi))
```

### Diferenças Absolutas

```python
soma_difs_cdf_normal=0
soma_difs_cdf_semi=0

for i in range(1,len(cdf_normal)-1):
    soma_difs_cdf_normal = abs(freq_acum[i] - cdf_normal[i])
for k in range(1,len(cdf_semicircular)-1):
    soma_difs_cdf_semi = abs(freq_acum[k] - cdf_semicircular[k])
    
print("Erro da soma do módulo das diferenças da normal versus F.R.A: {0} \nErro da soma do módulo das diferenças da semicircular versus F.R.A: {1}".format(soma_difs_cdf_normal, soma_difs_cdf_semi))
```

### Teste de aderência de Kolmogorov-Smirnov

```python
ks_normal = max(abs(freq_acum-cdf_normal))
ks_semi = max(abs(freq_acum-cdf_semicircular))
media=quant8_9.mean()

print("Teste K-S para a normal= {:4.2f}: {:4.3f}\n\
Teste K-S para a semicircular com scale={:4.3f}, loc=0 e a={:4.3f}: {:4.3f}".format(media, ks_normal, cfit[0], cfit[2], ks_semi))
```

De acordo com os dados calculados e apresentados na tabela, fica claro que a distribuição <strong>Normal</strong> representa melhor os dados analisados, pois obteve em todos os testes, um valor de erro menor que <strong>semicircular</strong>, portanto, uma aproximação maior ao valor real dos dados.
