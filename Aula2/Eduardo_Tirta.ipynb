```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
PNAD= pd.read_excel('PNAD_Eduardo_Tirta.xlsx')
```

```python
PNAD.shape
```

```python
n=PNAD.shape[0]
```

```python
list(PNAD)
```

```python
PNAD.UF = PNAD.UF.astype('category')
PNAD.SEXO = PNAD.SEXO.astype('category')
PNAD.IDADE = PNAD.IDADE.astype('category')
PNAD.RAÇA = PNAD.RAÇA.astype('category')
PNAD.REDEDEENSINO=PNAD.REDEDEENSINO.astype('category')
PNAD.INTERNET = PNAD.INTERNET.astype('category')
PNAD.NIVELDEENSINO = PNAD.NIVELDEENSINO.astype('category')
```

```python
# Tabelas univariada
print('Frequências de Pessoas de cada UF:')
ut1 = (PNAD.UF.value_counts()/n*100).round(decimals=1).reindex(['Norte','Nordeste','Sudeste','Sul','Centro Oeste'])
print(ut1,'\n')

print('Frequências relativas do Sexo:')
ut2 = (PNAD.SEXO.value_counts()/n*100).round(decimals=1).reindex(['Masculino','Feminino'])
print(ut2,'\n')

print('Frequências relativas da Idade:')
ut3 = (PNAD.IDADE.value_counts()/n*100).round(decimals=1).reindex(['Adolescente','Adulto'])
print(ut3,'\n')

print('Frequências relativas de Raças:')
ut4 = (PNAD.RAÇA.value_counts()/n*100).round(decimals=1).reindex(['Branca','Preta','Amarela','Parda','Indigena'])
print(ut4,'\n')

print('Frequências relativas de Redes de Ensino:')
ut5 = (PNAD.REDEDEENSINO.value_counts()/n*100).round(decimals=1).reindex(['Publica','Privada'])
print(ut5,'\n')

print('Frequências relativas de Disponibilidade de Internet:')
ut6 = (PNAD.INTERNET.value_counts()/n*100).round(decimals=1).reindex(['Sim','Nao'])
print(ut6,'\n')

print('Frequências relativas do Nivel de Ensino:')
ut7 = (PNAD.NIVELDEENSINO.value_counts()/n*100).round(decimals=1).reindex(['Ensino Superior', 'Ensino Medio'])
print(ut7,'\n')
```

```python
# Gráfico de barras e pizza dos UF
plot = ut1.plot(kind='bar',title='Regioes',figsize=(6, 6),color=('b','g','r','c','m'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('UF')


```

```python

plot = ut1.plot(kind='pie',title='UF',figsize=(6, 6))
```

```python
plot = ut3.plot(kind='pie',title='Idade',autopct='%.1f',figsize=(6, 6))
```

```python
plot = ut4.plot(kind='pie',title='Raças',autopct='%.1f',figsize=(6, 6))
```

```python
plot = ut5.plot(kind='pie',title='Rede de Ensino',autopct='%.1f',figsize=(6, 6))
```

```python
# Gráfico de barras e pizza da Disponibilidade de Internet
plot = ut6.plot(kind='bar',title='Idade',figsize=(6, 6),color=('b','r'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Disponibilidade de Internet')
```

```python
plot = ut6.plot(kind='pie',title='Disponibilidade de Internet',autopct='%.1f',figsize=(6, 6))
```

```python
#Tabela cruzada UF x Rede de Ensino
print('Tabela cruzada UF x Rede de Ensino:')
ct=pd.crosstab(PNAD.UF,PNAD.REDEDEENSINO,margins=True,rownames=['UF'], colnames=['REDE DE ENSINO'])
ct=ct.loc[['Norte','Nordeste','Sudeste','Sul','Centro Oeste']]
ct
```

```python
ct1 = pd.crosstab(PNAD.UF,PNAD.REDEDEENSINO,margins=True,rownames=['UF'], colnames=['REDE DE ENSINO'])
ct1 = ct1.loc[['Norte','Nordeste','Sudeste','Sul','Centro Oeste']]
ct1 = (ct1/len(PNAD)*100).round(decimals=0)
ct1 = ct1.iloc[:]
plot = ct1.plot(kind='bar')
plt.ylabel('Freq.Relativa (em%)')
```

```python
#Tabela cruzada REDE DE ENSINO x NIVEL DE ENSINO
print('Tabela cruzada REDE DE ENSINO x NIVEL DE ENSINO:')
ct2=pd.crosstab(PNAD.REDEDEENSINO,PNAD.NIVELDEENSINO,margins=True,rownames=['REDE DE ENSINO'], colnames=['NIVEL DE ENSINO'])
ct2=ct2.loc[['Privada','Publica']]
ct2
```

```python
ct1 = pd.crosstab(PNAD.REDEDEENSINO,PNAD.NIVELDEENSINO,margins=True,rownames=['REDE DE ENSINO'], colnames=['NIVEL DE ENSINO'])
ct1 = ct1.loc[['Privada','Publica']]
ct1 = (ct1/len(PNAD)*100).round(decimals=0)
ct1 = ct1.iloc[0:3,]
plot = ct1.plot(kind='bar')
plt.ylabel('Freq.Relativa (em%)')
```

```python
#Tabela cruzada INTERNET x NIVEL DE ENSINO
print('Tabela cruzada INTERNET x NIVEL DE ENSINO:')
ct2=pd.crosstab(PNAD.INTERNET,PNAD.NIVELDEENSINO,margins=True,rownames=['INTERNET'], colnames=['NIVEL DE ENSINO'])
ct2=ct2.loc[['Sim','Nao']]
ct2
```

```python
ct1 = pd.crosstab(PNAD.INTERNET,PNAD.NIVELDEENSINO,margins=True,rownames=['INTERNET'], colnames=['NIVEL DE ENSINO'])
ct1 = ct1.loc[['Sim','Nao']]
ct1 = (ct1/len(PNAD)*100).round(decimals=0)
ct1 = ct1.iloc[0:2,]
plot = ct1.plot(kind='bar')
plt.ylabel('Freq.Relativa (em%)')
```

```python
#Tabela cruzada INTERNET x UF
print('Tabela cruzada INTERNET x UF:')
ct=pd.crosstab(PNAD.INTERNET,PNAD.UF,margins=True,rownames=['UF'], colnames=['REDE DE ENSINO'])
ct=ct.loc[['Sim','Nao']]
ct
```

```python
ct1 = pd.crosstab(PNAD.INTERNET,PNAD.UF,margins=True,rownames=['INTERNET'], colnames=['UF'])
ct1 = ct1.loc[['Sim','Nao']]
ct1 = (ct1/len(PNAD)*100).round(decimals=0)
ct1 = ct1.iloc[:]
plot = ct1.plot(kind='bar',legend=False)
plt.ylabel('Freq.Relativa (em%)')
```

```python
ct1 = pd.crosstab(PNAD.UF,PNAD.INTERNET,margins=True,rownames=['UF'], colnames=['INTERNET'])
ct1 = ct1.loc[['Norte','Nordeste','Sudeste','Sul','Centro Oeste']]
ct1 = (ct1/len(PNAD)*100).round(decimals=0)
ct1 = ct1.iloc[:]
plot = ct1.plot(kind='bar',stacked=True,title='Internet em cada Plano')
plt.ylabel('Freq.Relativa (em%)')
```
